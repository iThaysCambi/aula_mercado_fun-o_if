```python
# 1 - criar uma lista com nome de 5 pets
lista_pets = ['Lua', 'Banzé', 'Pingo', 'Tico', 'Funy']

# 2 - função para exibir todos os pets
def ver_pets():
    print('\n------ Lista de Pets ------')
    for pet in lista_pets:
        print(pet)

# 3 - função para cadastrar um novo pet
def adicionar_pets():
    novo_pet = input('Digite o nome do novo pet: ')
    lista_pets.append(novo_pet)
    print('Pet cadastrado com sucesso!')

# 4 - menu de opções
print('------ Exibir os Pets ------')
print('Escolha a opção: ')
print('1 - ver os pets')
print('2 - cadastrar novo pet')

opcao = int(input('Informe a opção desejada: '))

if opcao == 1:
    ver_pets()

elif opcao == 2:
    adicionar_pets()
    ver_pets()

else:
    print('Opção não encontrada')
```
